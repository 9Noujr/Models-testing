```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
from joblib import load
```

```python
drive_path = '/content/drive/My Drive/MODELLING/rf_optimized_modelA_CrVI.joblib'
modelA_CrVI = load(drive_path)
print("ModelA_CrVI loaded successfully.")
drive_path = '/content/drive/My Drive/MODELLING/rf_optimized_modelA_totalCr.joblib'
modelA_totalCr = load(drive_path)
print("ModelA_totalCr loaded successfully.")
drive_path = '/content/drive/My Drive/MODELLING/rf_optimized_modelB_CrVI.joblib'
modelB_CrVI = load(drive_path)
print("ModelB_CrVI loaded successfully.")
drive_path = '/content/drive/My Drive/MODELLING/rf_optimized_modelB_totalCr.joblib'
modelB_totalCr = load(drive_path)
print("ModelB_totalCr loaded successfully.")
```

```python
X_testA_1 = '/content/drive/MyDrive/MODELLING/X_testA_1.csv'
X_testA_2 = '/content/drive/MyDrive/MODELLING/X_testA_2.csv'
X_testB_1 = '/content/drive/MyDrive/MODELLING/X_testB_1.csv'
X_testB_2 = '/content/drive/MyDrive/MODELLING/X_testB_2.csv'
y_testA_1 = '/content/drive/MyDrive/MODELLING/y_testA_1.csv'
y_testA_2 = '/content/drive/MyDrive/MODELLING/y_testA_2.csv'
y_testB_1 = '/content/drive/MyDrive/MODELLING/y_testB_1.csv'
y_testB_2 = '/content/drive/MyDrive/MODELLING/y_testB_2.csv'

X_testA_CrVI = pd.read_csv(X_testA_1)
X_testA_totalCr = pd.read_csv(X_testA_2)
X_testB_CrVI = pd.read_csv(X_testB_1)
X_testB_totalCr = pd.read_csv(X_testB_2)
y_testA_CrVI = pd.read_csv(y_testA_1)
y_testA_totalCr = pd.read_csv(y_testA_2)
y_testB_CrVI = pd.read_csv(y_testB_1)
y_testB_totalCr = pd.read_csv(y_testB_2)
```

```python
modelA_CrVI_prediction = modelA_CrVI.predict (X_testA_CrVI)
modelB_CrVI_prediction = modelB_CrVI.predict (X_testB_CrVI)
modelA_totalCr_prediction = modelA_totalCr.predict (X_testA_totalCr)
modelB_totalCr_prediction = modelB_totalCr.predict (X_testB_totalCr)

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import r2_score

modelA_CrVI_MAE = mean_absolute_error(modelA_CrVI_prediction, y_testA_CrVI)
modelA_CrVI_MAPE = mean_absolute_percentage_error(modelA_CrVI_prediction, y_testA_CrVI)
modelA_CrVI_r2 = r2_score(modelA_CrVI_prediction, y_testA_CrVI)
modelB_CrVI_MAE = mean_absolute_error(modelB_CrVI_prediction, y_testB_CrVI)
modelB_CrVI_MAPE = mean_absolute_percentage_error(modelB_CrVI_prediction, y_testB_CrVI)
modelB_CrVI_r2 = r2_score(modelB_CrVI_prediction, y_testB_CrVI)
modelA_totalCr_MAE = mean_absolute_error(modelA_totalCr_prediction, y_testA_totalCr)
modelA_totalCr_MAPE = mean_absolute_percentage_error(modelA_totalCr_prediction, y_testA_totalCr)
modelA_totalCr_r2 = r2_score(modelA_totalCr_prediction, y_testA_totalCr)
modelB_totalCr_MAE = mean_absolute_error(modelB_totalCr_prediction, y_testB_totalCr)
modelB_totalCr_MAPE = mean_absolute_percentage_error(modelB_totalCr_prediction, y_testB_totalCr)
modelB_totalCr_r2 = r2_score(modelB_totalCr_prediction, y_testB_totalCr)

print ('MAE A_Cr', modelA_CrVI_MAE), print ('MAE A_totalCr', modelA_totalCr_MAE)
print ('MAPE A_Cr', modelA_CrVI_MAPE), print ('MAPE A_totalCr', modelA_totalCr_MAPE)
print ('R2 A_Cr', modelA_CrVI_r2), print ('R2 A_totalCr', modelA_totalCr_r2)


print ('MAE B_Cr', modelB_CrVI_MAE), print ('MAE B_totalCr', modelB_totalCr_MAE)
print ('MAPE B_Cr', modelB_CrVI_MAPE), print ('MAPE B_totalCr', modelB_totalCr_MAPE)
print ('R2 B_Cr', modelB_CrVI_r2), print ('R2 B_totalCr', modelB_totalCr_r2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(y_testA_CrVI, modelA_CrVI_prediction, color='blue', label='Predicted vs Actual (Model A)')
plt.scatter(y_testB_CrVI, modelB_CrVI_prediction, color='red', label='Predicted vs Actual (Model B)')
plt.plot([0, 25], [0, 25], color='green', linestyle='--', label='Perfect Prediction Line')
plt.title('Prediction vs Actual (Model A vs Model B) Cr(VI)')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(y_testA_totalCr, modelA_totalCr_prediction, color='blue', label='Predicted vs Actual (Model A)')
plt.scatter(y_testB_totalCr, modelB_totalCr_prediction, color='red', label='Predicted vs Actual (Model B)')
plt.plot([0, 25], [0, 25], color='green', linestyle='--', label='Perfect Prediction Line')
plt.title('Prediction vs Actual (Model A vs Model B) Total Cr')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.show()
```
